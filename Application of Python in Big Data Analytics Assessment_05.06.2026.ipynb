### 1. Write a program to display all odd numbers between a given range. [5]

```python
start = int(input('Enter the start of the range:'))
end = int(input('Enter the end of the range:'))
print('\nOdd number is the given range:')
for i in range(start, end+1):
    if i%2==1:
        print(i, end = ' ')
    else:
        continue
```

### 3. Write a program to store 10 random integers between 1 and 20 into a list. Display the list. [6]

```python
import random as rnd
my_list = []
count= 1
while count<=10:
    num = round((rnd.random()*100))
    if num>=1 and num<=20:
        my_list.append(num)
        count+=1
    else:
        continue
print(my_list)
```

```python
from random import randint
my_list = []
for i in range(1,10+1):
    my_list.append(randint(1,20))
print('List of Random Integers between 1 and 20:\n', my_list)
```

```python
my_list = [rnd.randint(1,20) for i in range(10)]
my_list
```

### 4. Find the highest and lowest integers in the list. [4]

```python
lowest = min(my_list)
highest = max(my_list)
print('Lowest:',lowest)
print('Highest:',highest)
```

### 12. Build a machine learning model that can predict whether or not a student is selected based on all the other variables except their ID. Training and testing sets should be split on a ratio of 75:25. Evaluate the performance of the model using an appropriate evaluation metric. [15]

```python
#Import libraries
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
#Select features
x = df.drop(columns = ['student_id', 'selected'])
y = df[['selected']]
```

```python
#Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```python
#Standardise the x feature
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#Train the model
knn = KNeighborsClassifier(n_neighbors = 101)
knn.fit(x_train, y_train)
knn
```

```python
#Test the model using the testing set
predicted_selection = knn.predict(x_test)
```

```python
#Evaluate the model
acc_score = accuracy_score(y_test, predicted_selection)
print('Model accuracy score:', acc_score)
```
